```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set_style("whitegrid")
```

```python
# Load the Titanic dataset
df = pd.read_csv("titanic.csv")

# Display first 5 rows
df.head()
```

## 1. Dataset Overview

In this section, we explore the basic structure of the Titanic dataset. We examine the number of rows and columns, column names, data types, and statistical summary to understand the dataset before performing detailed analysis.

```python
df.info()   
```

## 2. Statistical Summary

The describe() function provides important statistical information such as mean, minimum, maximum, median, and quartiles for numerical columns.

```python
df.describe()
```

## 3. Missing Values Analysis

Missing values are identified to understand data quality. Columns with missing values may require cleaning before building machine learning models.

```python
df.isnull().sum()
```

## 4. Value Counts Analysis

The value_counts() function is used to count the occurrences of unique values in categorical columns. This helps us understand the distribution of categories in the dataset.

```python
# Survival count
print("Survival Count:\n")
print(df["Survived"].value_counts())

print("\n")

# Gender count
print("Gender Count:\n")
print(df["Sex"].value_counts())

print("\n")

# Passenger Class count
print("Passenger Class Count:\n")
print(df["Pclass"].value_counts())
```

## 5. Age Distribution

Histogram is used to understand the distribution of passenger ages.

```python
plt.figure(figsize=(8,5))

plt.hist(df["Age"], bins=20)

plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")

plt.show()
```

### Observation

- Most passengers were between 20 and 40 years of age.
- Very few passengers were above 70 years old.
- The age distribution is slightly right-skewed.

## 6. Boxplot of Age

A boxplot is used to identify the spread of passenger ages and detect any outliers in the dataset.

```python
plt.figure(figsize=(8,5))

sns.boxplot(x=df["Age"])

plt.title("Boxplot of Age")
plt.xlabel("Age")

plt.show()
```

### Observation

- Most passengers are concentrated in the middle age range.
- A few older passengers appear as outliers.
- The median age lies approximately in the center of the box.

## 7. Survival Distribution

This count plot shows the number of passengers who survived and those who did not survive.

```python
plt.figure(figsize=(6,4))

sns.countplot(x="Survived", data=df)

plt.title("Survival Distribution")
plt.xlabel("Survival Status")
plt.ylabel("Number of Passengers")

plt.show()
```

### Observation

- The number of passengers who did not survive is higher than those who survived.
- The dataset is slightly imbalanced with respect to survival.

## 8. Gender Distribution

This count plot shows the distribution of male and female passengers.

```python
plt.figure(figsize=(6,4))

sns.countplot(x="Sex", data=df)

plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Number of Passengers")

plt.show()
```

### Observation

- Male passengers are more than female passengers.
- The dataset contains a higher proportion of males.

```python
plt.figure(figsize=(6,4))

sns.countplot(x="Pclass", data=df)

plt.title("Passenger Class Distribution")
plt.xlabel("Passenger Class")
plt.ylabel("Number of Passengers")

plt.show()
```

### Observation

- Most passengers belonged to the third class.
- First-class passengers were fewer compared to second and third class.

## 10. Scatter Plot: Age vs Fare

A scatter plot is used to understand the relationship between passenger age and ticket fare.

```python
plt.figure(figsize=(8,5))

sns.scatterplot(x="Age", y="Fare", data=df)

plt.title("Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")

plt.show()
```

### Observation

- There is no strong relationship between Age and Fare.
- Some passengers paid significantly higher fares regardless of age.
- A few fare values appear as outliers.

## 11. Correlation Heatmap

A heatmap helps visualize the correlation between numerical variables in the dataset.


```python
plt.figure(figsize=(8,6))

sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")

plt.title("Correlation Heatmap")

plt.show()
```

### Observation

- Passenger Class has a negative correlation with Fare.
- Fare has a slight positive relationship with Survival.
- Most variables have weak correlations with each other.

## 12. Pairplot

A pairplot visualizes relationships between multiple numerical variables simultaneously.

```python
sns.pairplot(df[["Age","Fare","Pclass","Survived"]])

plt.show()
```

### Observation

- Pairplot helps compare multiple numerical variables together.
- It shows data distribution along the diagonal.
- Scatter plots reveal relationships between different variables.
